```python

# -*- coding: utf-8 -*-
"""fShrink_HighNA_EUV_Stochastic_Benchmark.ipynb

# f-Shrink for High-NA EUV Stochastic Control Benchmark

This benchmark tests f-Shrink controllers against the most challenging EUV problems:
- High-NA (0.55) stochastic effects
- Photon shot noise
- Chemical amplification noise
- Mask 3D effects
- Thermal drift
- Multi-layer alignment
"""

!pip install -q numpy matplotlib scipy tqdm

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.integrate import solve_ivp
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
F = 0.4745409995126511

print("="*80)
print("f-Shrink for High-NA EUV Stochastic Control Benchmark")
print("="*80)

# ==================== 1. Advanced EUV Stochastic Model ====================
print("\n" + "="*80)
print("1. High-NA EUV Stochastic Model (0.55 NA)")
print("="*80)

class HighNAEUVModel:
    """
    Advanced High-NA EUV stochastic model with:
    - Photon shot noise (Poisson)
    - Chemical amplification noise
    - Mask 3D phase effects
    - Thermal deformation
    - Multi-layer alignment drift
    - Stochastic failure modes (micro-bridging, line breaks)
    """

    def __init__(self, dt=1e-4, mode='high_na'):
        self.dt = dt
        self.t = 0.0
        self.mode = mode  # 'high_na' or 'extreme'

        # ===== High-NA parameters (0.55 NA) =====
        if mode == 'high_na':
            self.NA = 0.55
            self.photon_flux = 100  # photons/nm² (low for High-NA)
            self.k1 = 0.3  # process factor
        else:  # extreme case for stress testing
            self.NA = 0.55
            self.photon_flux = 50  # extremely low photons
            self.k1 = 0.25

        self.wavelength = 13.5  # nm (EUV)
        self.resolution = self.k1 * self.wavelength / self.NA  # Rayleigh criterion

        # ===== Stochastic parameters =====
        self.photon_noise_floor = 0.05  # nm
        self.chemical_noise = 0.1  # nm
        self.mask_3d_amplitude = 0.3  # nm
        self.mask_3d_freq = 1000  # Hz

        # ===== Thermal parameters =====
        self.thermal_capacity = 1000  # J/K
        self.thermal_resistance = 0.01  # K/W
        self.thermal_time_constant = 10.0  # s
        self.base_temp = 22.0  # °C
        self.temp = self.base_temp

        # ===== Mechanical parameters =====
        self.stage_mass = 500  # kg
        self.stage_stiffness = 1e7  # N/m
        self.stage_damping = 1e5  # N/(m/s)
        self.gravity = 9.8  # m/s²

        # ===== Multi-layer alignment =====
        self.n_layers = 5
        self.layer_thickness = 30  # nm
        self.overlay_budget = 1.5  # nm (3nm node requirement)
        self.alignment_noise = 0.2  # nm per layer

        # ===== Failure thresholds =====
        self.micro_bridge_threshold = 2.0  # nm (CD variation causing bridges)
        self.line_break_threshold = 1.5  # nm (CD variation causing breaks)

        # ===== State indices =====
        self.idx_x = 0      # position (nm)
        self.idx_v = 1      # velocity (nm/s)
        self.idx_f = 2      # focus (nm)
        self.idx_d = 3      # dose (mJ/cm²)
        self.idx_cd = 4     # critical dimension (nm)
        self.idx_ovl = 5    # overlay error (nm)
        self.idx_temp = 6   # temperature (°C)
        self.n_states = 7

        # State limits
        self.state_limits = np.array([[-20, 20],      # x
                                       [-1000, 1000],  # v
                                       [-20, 20],      # f
                                       [0, 50],        # d
                                       [10, 30],       # cd
                                       [-3, 3],        # ovl
                                       [20, 30]])      # temp

        # History for stochastic analysis
        self.failure_history = []
        self.cd_history = []

    def photon_shot_noise(self, dose):
        """Photon shot noise (Poisson statistics)"""
        # Number of photons per area
        n_photons = dose * self.photon_flux * 10  # scale factor
        if n_photons <= 0:
            return 0
        # Poisson noise standard deviation = sqrt(N)
        noise_std = np.sqrt(n_photons) / n_photons * self.photon_noise_floor
        return np.random.randn() * noise_std

    def chemical_amplification_noise(self, dose):
        """Chemical amplification noise (nonlinear)"""
        base_noise = self.chemical_noise * (1 + 0.1 * np.random.randn())
        # Nonlinear amplification at high dose
        if dose > 30:
            base_noise *= 1.5
        return base_noise * np.random.randn()

    def mask_3d_effect(self):
        """Mask 3D phase effects (time-varying)"""
        phase = 2 * np.pi * self.mask_3d_freq * self.t
        return self.mask_3d_amplitude * np.sin(phase)

    def thermal_dynamics(self, power):
        """Thermal dynamics"""
        dtemp = (power * 1e-6 - (self.temp - self.base_temp) /
                 self.thermal_time_constant) / self.thermal_capacity
        return dtemp * self.dt

    def mechanical_dynamics(self, u_x, x, v):
        """Stage mechanical dynamics with gravity compensation"""
        spring_force = -self.stage_stiffness * x
        damping_force = -self.stage_damping * v
        control_force = u_x * 1000  # N
        gravity_force = self.stage_mass * self.gravity

        total_force = control_force + spring_force + damping_force - gravity_force
        accel = total_force / self.stage_mass

        dx = v * self.dt
        dv = accel * self.dt

        return dx, dv

    def stochastic_failure_detection(self, cd, dose, overlay):
        """Detect stochastic failures"""
        failures = []

        # Micro-bridging (CD too large)
        if cd > self.micro_bridge_threshold + 15:  # base CD 15nm
            failures.append('micro_bridge')

        # Line break (CD too small)
        if cd < 15 - self.line_break_threshold:
            failures.append('line_break')

        # Overlay failure
        if abs(overlay) > self.overlay_budget:
            failures.append('overlay_error')

        # Dose-related failure
        if dose < 10 or dose > 45:
            failures.append('dose_out_of_range')

        return failures

    def dynamics(self, state, u_x, u_f, u_d):
        """Full nonlinear stochastic dynamics"""
        x, v, f, d, cd, ovl, temp = state

        # ===== Control inputs with limits =====
        u_x = np.clip(u_x, -20, 20)
        u_f = np.clip(u_f, -20, 20)
        u_d = np.clip(u_d, -20, 20)

        # ===== Power calculation for thermal =====
        power = u_x**2 + u_f**2 + u_d**2

        # ===== Stochastic noise sources =====
        shot_noise = self.photon_shot_noise(d)
        chem_noise = self.chemical_amplification_noise(d)
        mask_effect = self.mask_3d_effect()

        # ===== Mechanical dynamics =====
        dx, dv = self.mechanical_dynamics(u_x, x, v)

        # ===== Focus dynamics (with thermal drift) =====
        thermal_drift = 0.01 * (temp - self.base_temp)
        df = (u_f * 2 - 5 * f + thermal_drift + mask_effect) * self.dt * 10

        # ===== Dose dynamics =====
        dd = (u_d * 2 - 0.1 * d) * self.dt * 10

        # ===== CD evolution (nonlinear, stochastic) =====
        target_cd = 15.0  # nm target
        dose_effect = -0.5 * (d - 25) / 25  # dose response
        focus_effect = 0.1 * f**2  # focus nonlinearity
        stochastic_total = shot_noise + chem_noise

        cd_error = target_cd - cd
        dcd = (0.1 * cd_error + dose_effect + focus_effect + stochastic_total) * self.dt * 100
        cd = cd + dcd
        cd = np.clip(cd, 10, 30)

        # ===== Overlay error (multi-layer alignment) =====
        # Each layer contributes random offset
        layer_errors = np.random.randn(self.n_layers) * self.alignment_noise
        ovl_error = np.sum(layer_errors) + 0.01 * x + 0.005 * self.t
        ovl = ovl_error

        # ===== Temperature update =====
        temp += self.thermal_dynamics(power)
        temp = np.clip(temp, 20, 30)

        # ===== State update =====
        new_state = np.array([
            x + dx,
            v + dv,
            f + df,
            d + dd,
            cd,
            ovl,
            temp
        ])

        # Apply state limits
        for i in range(self.n_states):
            new_state[i] = np.clip(new_state[i],
                                   self.state_limits[i, 0],
                                   self.state_limits[i, 1])

        self.t += self.dt

        # Detect failures
        failures = self.stochastic_failure_detection(cd, d, ovl)
        if failures:
            self.failure_history.append({
                'time': self.t,
                'failures': failures,
                'state': new_state.copy()
            })

        return new_state

    def measure(self, state):
        """Realistic measurements with noise and quantization"""
        x, v, f, d, cd, ovl, temp = state

        # Position measurement (nm resolution encoder)
        x_meas = np.round(x / 0.1) * 0.1 + 0.02 * np.random.randn()

        # Focus measurement (interferometer)
        f_meas = f + 0.05 * np.random.randn()

        # Dose measurement (power monitor)
        d_meas = d + 0.1 * np.random.randn()

        # CD measurement (SEM, destructive sampling - not available in real-time)
        # In real system, CD is measured after exposure
        cd_meas = cd + 0.2 * np.random.randn()

        return x_meas, f_meas, d_meas, cd_meas

# ==================== 2. Controllers ====================
print("\n" + "="*80)
print("2. Advanced Controllers")
print("="*80)

class BaseController:
    def __init__(self, dt, setpoint=0.0, name='Base'):
        self.dt = dt
        self.setpoint = setpoint
        self.name = name
        self.reset()

    def reset(self):
        self.integral = 0.0
        self.prev_error = 0.0
        self.kp = 5.0
        self.ki = 1.0
        self.kd = 0.5
        self.integral_limit = 100.0

    def update_gains(self):
        pass

    def control(self, measurement):
        self.update_gains()
        error = self.setpoint - measurement

        self.integral += error * self.dt
        self.integral = np.clip(self.integral, -self.integral_limit, self.integral_limit)

        derivative = (error - self.prev_error) / (self.dt + 1e-9)
        self.prev_error = error

        output = self.kp * error + self.ki * self.integral + self.kd * derivative
        return np.clip(output, -20, 20)

class PIDController(BaseController):
    def __init__(self, dt, setpoint=0.0):
        super().__init__(dt, setpoint, 'PID')

    def update_gains(self):
        pass

def safe_pow(base, exp):
    if base <= 0:
        return 0
    exp_clipped = np.clip(exp, -20, 20)
    return base ** exp_clipped

def f_shrink_fast(x, limit=20.0):
    if abs(x) < 1e-12:
        return 0.0
    sign = np.sign(x)
    abs_x = np.abs(x)
    exp_arg = abs_x / (2 * F)
    exp_arg = np.clip(exp_arg, -20, 20)
    pow_val = safe_pow(F, exp_arg)
    result = sign * np.sqrt(abs_x) * pow_val
    return np.clip(result, -limit, limit)

def f_shrink_decay(x, beta, target=F, limit=20.0):
    result = (1 - beta) * x + beta * target
    return np.clip(result, -limit, limit)

def f_shrink_smooth(x, alpha, limit=20.0):
    target = f_shrink_fast(x, limit=limit*2)
    result = (1 - alpha) * x + alpha * target
    return np.clip(result, -limit, limit)

class FastController(BaseController):
    def __init__(self, dt, setpoint=0.0):
        super().__init__(dt, setpoint, 'Fast')

    def update_gains(self):
        self.kp = f_shrink_fast(self.kp)
        self.ki = f_shrink_fast(self.ki)
        self.kd = f_shrink_fast(self.kd)

class DecayController(BaseController):
    def __init__(self, dt, beta, setpoint=0.0):
        super().__init__(dt, setpoint, f'Decay β={beta}')
        self.beta = beta

    def update_gains(self):
        self.kp = f_shrink_decay(self.kp, self.beta)
        self.ki = f_shrink_decay(self.ki, self.beta)
        self.kd = f_shrink_decay(self.kd, self.beta)

class SmoothController(BaseController):
    def __init__(self, dt, alpha, setpoint=0.0):
        super().__init__(dt, setpoint, f'Smooth α={alpha}')
        self.alpha = alpha

    def update_gains(self):
        self.kp = f_shrink_smooth(self.kp, self.alpha)
        self.ki = f_shrink_smooth(self.ki, self.alpha)
        self.kd = f_shrink_smooth(self.kd, self.alpha)

# ==================== 3. Benchmark Simulation ====================
print("\n" + "="*80)
print("3. Running High-NA EUV Benchmark")
print("="*80)

def run_euv_benchmark(controller_type, params=None, T=5.0, mode='high_na'):
    """Run High-NA EUV stochastic benchmark"""
    dt = 2e-4
    n_steps = int(T / dt)
    system = HighNAEUVModel(dt, mode)

    # Initialize controllers for each axis
    if controller_type == 'PID':
        ctrl_x = PIDController(dt, setpoint=0.0)
        ctrl_f = PIDController(dt, setpoint=0.0)
        ctrl_d = PIDController(dt, setpoint=25.0)  # target dose 25 mJ/cm²
        name = 'PID'
    elif controller_type == 'fast':
        ctrl_x = FastController(dt, setpoint=0.0)
        ctrl_f = FastController(dt, setpoint=0.0)
        ctrl_d = FastController(dt, setpoint=25.0)
        name = 'Fast'
    elif controller_type == 'decay':
        beta = params['beta']
        ctrl_x = DecayController(dt, beta, setpoint=0.0)
        ctrl_f = DecayController(dt, beta, setpoint=0.0)
        ctrl_d = DecayController(dt, beta, setpoint=25.0)
        name = f'Decay β={beta}'
    elif controller_type == 'smooth':
        alpha = params['alpha']
        ctrl_x = SmoothController(dt, alpha, setpoint=0.0)
        ctrl_f = SmoothController(dt, alpha, setpoint=0.0)
        ctrl_d = SmoothController(dt, alpha, setpoint=25.0)
        name = f'Smooth α={alpha}'

    # Initial state
    state = np.array([5.0,      # x (nm offset)
                      0.0,      # v
                      2.0,      # f (focus error)
                      20.0,     # d (dose)
                      15.0,     # cd (target)
                      0.0,      # ovl (overlay)
                      23.0])    # temp (°C)

    # Data recording
    time = np.arange(n_steps) * dt
    x_hist = np.zeros(n_steps)
    f_hist = np.zeros(n_steps)
    d_hist = np.zeros(n_steps)
    cd_hist = np.zeros(n_steps)
    ovl_hist = np.zeros(n_steps)
    epe_hist = np.zeros(n_steps)
    failures = []
    gains_hist = []

    for i in tqdm(range(n_steps), desc=f'Running {name}', leave=False):
        # Get measurements
        x_meas, f_meas, d_meas, cd_meas = system.measure(state)

        # Control computation
        u_x = ctrl_x.control(x_meas)
        u_f = ctrl_f.control(f_meas)
        u_d = ctrl_d.control(25.0 - d_meas)  # dose control

        # System dynamics
        state = system.dynamics(state, u_x, u_f, u_d)

        # Record data
        x_hist[i] = state[system.idx_x]
        f_hist[i] = state[system.idx_f]
        d_hist[i] = state[system.idx_d]
        cd_hist[i] = state[system.idx_cd]
        ovl_hist[i] = state[system.idx_ovl]

        # EPE calculation (Edge Placement Error)
        epe_hist[i] = np.sqrt(x_hist[i]**2 + f_hist[i]**2) + 0.5 * np.abs(ovl_hist[i])

        # Record gains periodically
        if i % 500 == 0:
            gains_hist.append({
                'step': i,
                'kp_x': ctrl_x.kp,
                'ki_x': ctrl_x.ki,
                'kd_x': ctrl_x.kd
            })

    # Collect failures from system
    failures = system.failure_history

    return {
        'name': name,
        'time': time,
        'x': x_hist,
        'f': f_hist,
        'd': d_hist,
        'cd': cd_hist,
        'ovl': ovl_hist,
        'epe': epe_hist,
        'failures': failures,
        'gains': gains_hist
    }

# ==================== 4. Run All Controllers ====================
print("\nRunning High-NA EUV benchmark simulations...")

configs = [
    ('PID', None),
    ('fast', None),
    ('decay', {'beta': 0.02}),
    ('decay', {'beta': 0.035}),
    ('smooth', {'alpha': 0.05}),
    ('smooth', {'alpha': 0.1}),
    ('smooth', {'alpha': 0.15})
]

results = {}
for ctype, params in tqdm(configs, desc='Controllers'):
    name = ctype
    if params:
        if 'beta' in params:
            name += f"_beta_{params['beta']}"
        elif 'alpha' in params:
            name += f"_alpha_{params['alpha']}"
    print(f"\nTesting {name}...")
    results[name] = run_euv_benchmark(ctype, params, T=3.0, mode='high_na')

# ==================== 5. Performance Analysis ====================
print("\n" + "="*80)
print("4. Performance Analysis")
print("="*80)

performance = {}
for name, data in results.items():
    # Steady-state metrics (last 20%)
    n = len(data['epe'])
    steady_start = int(n * 0.8)

    steady_epe = np.mean(data['epe'][steady_start:])
    steady_x = np.mean(np.abs(data['x'][steady_start:]))
    steady_f = np.mean(np.abs(data['f'][steady_start:]))
    steady_ovl = np.mean(np.abs(data['ovl'][steady_start:]))

    # Transient metrics
    max_epe = np.max(data['epe'])
    settling_idx = np.where(np.abs(data['epe']) < 2.0)[0]
    settling_time = data['time'][settling_idx[0]] if len(settling_idx) > 0 else 3.0

    # CD control
    cd_error = np.mean(np.abs(data['cd'][steady_start:] - 15.0))
    cd_std = np.std(data['cd'][steady_start:])

    # Failure rate
    n_failures = len(data['failures'])
    failure_rate = n_failures / (data['time'][-1] * 1000)  # failures per ms

    performance[name] = {
        'steady_epe': steady_epe,
        'steady_x': steady_x,
        'steady_f': steady_f,
        'steady_ovl': steady_ovl,
        'max_epe': max_epe,
        'settling_time': settling_time,
        'cd_error': cd_error,
        'cd_std': cd_std,
        'failure_rate': failure_rate,
        'n_failures': n_failures
    }

# Sort by steady EPE
sorted_names = sorted(performance.keys(),
                     key=lambda x: performance[x]['steady_epe'])

print("\n" + "="*80)
print("High-NA EUV Benchmark Results (sorted by steady EPE)")
print("="*80)
print(f"{'Controller':<20} {'Steady EPE':<12} {'Max EPE':<12} "
      f"{'Settle(ms)':<12} {'CD Error':<12} {'Failures':<10}")
print("-"*90)

for name in sorted_names:
    p = performance[name]
    print(f"{name:<20} {p['steady_epe']:<12.3f} {p['max_epe']:<12.3f} "
          f"{p['settling_time']*1000:<12.1f} {p['cd_error']:<12.3f} "
          f"{p['n_failures']:<10}")

# ==================== 6. Visualization ====================
print("\n" + "="*80)
print("5. Generating Visualization")
print("="*80)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('High-NA EUV Stochastic Control Benchmark', fontsize=14)

# Select top controllers for visualization
top_names = sorted_names[:4]  # Top 4

# 1. EPE time series
ax = axes[0, 0]
for name in top_names:
    if name in results:
        ax.plot(results[name]['time']*1000, results[name]['epe'],
                label=name, linewidth=1)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('EPE (nm)')
ax.set_title('Edge Placement Error')
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_ylim(0, 10)

# 2. CD control
ax = axes[0, 1]
for name in top_names:
    if name in results:
        ax.plot(results[name]['time']*1000, results[name]['cd'],
                label=name, linewidth=1)
ax.axhline(y=15, color='r', linestyle='--', label='Target CD')
ax.set_xlabel('Time (ms)')
ax.set_ylabel('CD (nm)')
ax.set_title('Critical Dimension Control')
ax.legend()
ax.grid(True, alpha=0.3)

# 3. Overlay error
ax = axes[0, 2]
for name in top_names:
    if name in results:
        ax.plot(results[name]['time']*1000, results[name]['ovl'],
                label=name, linewidth=1)
ax.axhline(y=1.5, color='r', linestyle='--', label='Budget')
ax.axhline(y=-1.5, color='r', linestyle='--')
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Overlay (nm)')
ax.set_title('Overlay Error')
ax.legend()
ax.grid(True, alpha=0.3)

# 4. Steady EPE comparison
ax = axes[1, 0]
names = list(performance.keys())
steady_epe = [performance[n]['steady_epe'] for n in names]
bars = ax.bar(range(len(names)), steady_epe)
ax.set_xlabel('Controller')
ax.set_ylabel('Steady EPE (nm)')
ax.set_title('Steady-State EPE Comparison')
ax.set_xticks(range(len(names)))
ax.set_xticklabels(names, rotation=45, ha='right')
ax.grid(True, alpha=0.3, axis='y')
best_idx = np.argmin(steady_epe)
bars[best_idx].set_color('red')

# 5. Failure rate comparison
ax = axes[1, 1]
failures = [performance[n]['n_failures'] for n in names]
bars = ax.bar(range(len(names)), failures)
ax.set_xlabel('Controller')
ax.set_ylabel('Number of Failures')
ax.set_title('Stochastic Failures')
ax.set_xticks(range(len(names)))
ax.set_xticklabels(names, rotation=45, ha='right')
ax.grid(True, alpha=0.3, axis='y')
best_idx = np.argmin(failures)
bars[best_idx].set_color('red')

# 6. CD control comparison
ax = axes[1, 2]
cd_errors = [performance[n]['cd_error'] for n in names]
bars = ax.bar(range(len(names)), cd_errors)
ax.set_xlabel('Controller')
ax.set_ylabel('CD Error (nm)')
ax.set_title('Critical Dimension Error')
ax.set_xticks(range(len(names)))
ax.set_xticklabels(names, rotation=45, ha='right')
ax.grid(True, alpha=0.3, axis='y')
best_idx = np.argmin(cd_errors)
bars[best_idx].set_color('red')

plt.tight_layout()
plt.savefig('high_na_euv_benchmark_results.png', dpi=150)
plt.show()

# ==================== 7. Conclusions ====================
print("\n" + "="*80)
print("6. Conclusions")
print("="*80)

best_name = sorted_names[0]
best_perf = performance[best_name]
second_name = sorted_names[1]
second_perf = performance[second_name]

print(f"\n🏆 Best Controller: {best_name}")
print(f"   Steady EPE: {best_perf['steady_epe']:.3f} nm")
print(f"   Max EPE: {best_perf['max_epe']:.3f} nm")
print(f"   Settling Time: {best_perf['settling_time']*1000:.1f} ms")
print(f"   CD Error: {best_perf['cd_error']:.3f} nm")
print(f"   Failures: {best_perf['n_failures']}")

print(f"\n🥈 Second Place: {second_name}")
print(f"   Steady EPE: {second_perf['steady_epe']:.3f} nm")
print(f"   Improvement over PID: {(performance['PID']['steady_epe'] - best_perf['steady_epe']):.3f} nm")

print("\n✅ High-NA EUV Benchmark Complete")
print("   Results saved to 'high_na_euv_benchmark_results.png'")
```
